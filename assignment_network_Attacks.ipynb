```python
import pandas as pd
import numpy as np


data=pd.read_csv("IDS_train.csv")
df=pd.DataFrame(data)
print(data.info())
```

```python
data.shape
```

```python
data.replace([np.inf,-np.inf],np.nan,inplace=True)
```

```python
data.dropna(inplace=True)
```

```python
data.shape
```

```python
data["Label"].value_counts()
```

```python
sampled_data=data.sample(frac=0.08,random_state=1)
sampled_data.shape
```

```python
sampled_data.replace(to_replace=["DoS Hulk"],value="Attack",inplace=True)
```

```python
sampled_data.replace(to_replace=["DoS GoldenEye"],value="Attack",inplace=True)
```

```python
sampled_data.replace(to_replace=["DoS Slowhttptest"],value="Attack",inplace=True)
sampled_data.replace(to_replace=["DoS slowloris"],value="Attack",inplace=True)
```

```python
x=sampled_data
x.replace(to_replace="BENIGN",value=0,inplace=True)
x.replace(to_replace="Attack",value=1,inplace=True)
```

```python
from sklearn.model_selection import train_test_split
train,test= train_test_split(x,test_size=0.3,random_state=12)
print(train.shape)
```
